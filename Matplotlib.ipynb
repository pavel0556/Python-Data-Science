### Задача_2

```python
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f)
plt.title('График f(t)',fontsize=16,color='Green')

plt.xlabel('Значение t')
plt.ylabel('Значение f')

plt.axis([0.5,9.5,-2.5,2.5])
plt.show()
```
